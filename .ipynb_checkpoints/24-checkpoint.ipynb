```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("user_behavior_dataset.csv")
df.head()
```

```python
df = df.drop(['User ID'], axis=1)

# 4.2 Encode categorical variables: Device Model, Operating System, Gender
le_device = LabelEncoder()
le_os = LabelEncoder()
le_gender = LabelEncoder()

df['Device Model'] = le_device.fit_transform(df['Device Model'])
df['Operating System'] = le_os.fit_transform(df['Operating System'])
df['Gender'] = le_gender.fit_transform(df['Gender'])  # Male=1, Female=0
```

```python
# 4.3 Encode target column (User Behavior Class)
le_behavior = LabelEncoder()
df['User Behavior Class'] = le_behavior.fit_transform(df['User Behavior Class'])

print("\nâœ… Encoding Done Successfully!")
df.head()
```

```python
X = df.drop('User Behavior Class', axis=1)
y = df['User Behavior Class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nâœ… Feature Scaling Completed!")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
print(f"\nTraining Data Shape: {X_train.shape}")
print(f"Testing Data Shape: {X_test.shape}")
```

```python
kernels = ['linear', 'poly', 'rbf']
accuracies = {}

for kernel in kernels:
    print(f"\nðŸ”¹ Training SVM with {kernel.upper()} kernel...")
    model = SVC(kernel=kernel, C=1.0, gamma='scale', random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    accuracies[kernel] = acc * 100

    print(f"Accuracy ({kernel} kernel): {acc*100:.2f}%")
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

```python
print("\nðŸ“Š Accuracy Comparison (SVM Kernels):")
for k, v in accuracies.items():
    print(f"{k.upper()} Kernel: {v:.2f}%")

plt.figure(figsize=(6, 4))
plt.bar(accuracies.keys(), accuracies.values(), color=['skyblue', 'orange', 'green'])
plt.title("SVM Kernel Accuracy Comparison - User Behavior Classification")
plt.xlabel("Kernel Type")
plt.ylabel("Accuracy (%)")
plt.show()
```

```python
print("\nðŸ”¹ Test Model for a New User Input")

# Example input (values must follow dataset order)
# Device Model, Operating System, App Usage Time, Screen On Time, Battery Drain, Number of Apps, Data Usage, Age, Gender
sample = np.array([[le_device.transform(['Google Pixel 5'])[0],
                    le_os.transform(['Android'])[0],
                    300, 5.5, 1500, 55, 1000, 28, le_gender.transform(['Male'])[0]]])

# Scale the input
sample_scaled = scaler.transform(sample)

# Use best-performing model (usually RBF)
best_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
best_model.fit(X_train, y_train)

pred = best_model.predict(sample_scaled)
predicted_class = le_behavior.inverse_transform(pred)[0]

print(f"\nðŸŸ¢ Predicted User Behavior Class: {predicted_class}")
```
