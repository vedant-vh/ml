```python
# -----------------------------------------------------------
# K-MEANS CLUSTERING (Manual Centroids)
# -----------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define Data Points
points = {
    'P1': [0.1, 0.6],
    'P2': [0.15, 0.71],
    'P3': [0.08, 0.9],
    'P4': [0.16, 0.85],
    'P5': [0.2, 0.3],
    'P6': [0.25, 0.5],
    'P7': [0.24, 0.1],
    'P8': [0.3, 0.2]
}

df = pd.DataFrame(points).T
df.columns = ['x', 'y']

# Step 2: Initialize centroids
m1 = np.array([0.1, 0.6])   # P1
m2 = np.array([0.3, 0.2])   # P8

# Step 3: Assign each point to nearest centroid
def distance(p, c):
    return np.sqrt(np.sum((p - c) ** 2))

clusters = []
for i in range(len(df)):
    d1 = distance(df.iloc[i], m1)
    d2 = distance(df.iloc[i], m2)
    cluster = 1 if d1 < d2 else 2
    clusters.append(cluster)

df['Cluster'] = clusters

# Step 4: Print cluster assignments
print("Initial Cluster Assignment:\n")
print(df)

# Step 5: Find new centroids
C1 = df[df['Cluster'] == 1]
C2 = df[df['Cluster'] == 2]

m1_new = [C1['x'].mean(), C1['y'].mean()]
m2_new = [C2['x'].mean(), C2['y'].mean()]

# Step 6: Print required answers
print("\n1️⃣ P6 belongs to Cluster:", df.loc['P6', 'Cluster'])
print("2️⃣ Population of Cluster around m2:", len(C2))
print("3️⃣ Updated Centroids:")
print("   m1 =", tuple(np.round(m1_new, 3)))
print("   m2 =", tuple(np.round(m2_new, 3)))

# Step 7: Visualize the clusters
plt.figure(figsize=(6,5))
plt.scatter(C1['x'], C1['y'], color='red', label='Cluster 1')
plt.scatter(C2['x'], C2['y'], color='blue', label='Cluster 2')
plt.scatter(m1[0], m1[1], color='black', marker='*', s=200, label='Initial m1')
plt.scatter(m2[0], m2[1], color='purple', marker='*', s=200, label='Initial m2')
plt.scatter(m1_new[0], m1_new[1], color='orange', marker='X', s=200, label='New m1')
plt.scatter(m2_new[0], m2_new[1], color='green', marker='X', s=200, label='New m2')
plt.title('K-Means Clustering (Iteration 1)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
